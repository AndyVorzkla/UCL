```python
attacking = pd.read_csv('./data/attacking.csv', index_col=0)
attempts = pd.read_csv('./data/attempts.csv', index_col=0)
defending = pd.read_csv('./data/defending.csv', index_col=0)
disciplinary = pd.read_csv('./data/disciplinary.csv', index_col=0)
distributon = pd.read_csv('./data/distributon.csv', index_col=0)
goalkeeping = pd.read_csv('./data/goalkeeping.csv', index_col=0)
goals = pd.read_csv('./data/goals.csv', index_col=0)
key_stats = pd.read_csv('./data/key_stats.csv', index_col=0)
```

```python
def generator_help_merge(general_data: pd.DataFrame, list_of_data: list, column_on_merge: str):
    general_data = general_data.copy()
    general_columns = general_data.columns
    
    for data in list_of_data:
        columns = data.columns
        columns = [*filter(lambda x: x not in general_columns, columns)]
        general_data = general_data.merge(right=data[columns], how='left',  on=column_on_merge)
    else:
        return general_data
    
    
    
```

```python
all_data = generator_help_merge(key_stats, [attacking, attempts, goals, defending, disciplinary, goalkeeping, distributon], 'player_name')
```
